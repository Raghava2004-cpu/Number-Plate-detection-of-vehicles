```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt



number_plate_cascades = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_russian_plate_number.xml")

image = cv.imread("C:/Users/ragha/Downloads/images.jpeg")

grey = cv.cvtColor(image , cv.COLOR_BGR2GRAY)


full_body = number_plate_cascades.detectMultiScale(grey , 1.1 , 3)

for (x , y, w, h) in full_body:
    cv.putText(image , "Number Plate" , (x , y-10) , cv.FONT_HERSHEY_SIMPLEX , 0.5,(0,255,0) , 2)
    cv.rectangle(image , (x,y) , (x+w, y+h) , (255,0,0) , 3)

image = cv.resize(image , (500,400))
cv.imshow("Image" , image)
cv.waitKey(0)
cv.destroyAllWindows()  

plt.imshow(cv.cvtColor(image , cv.COLOR_BGR2RGB))
plt.axis("OFF")
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

plate_cascades =  cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_russian_plate_number.xml" )
video =  cv.VideoCapture("C:/Users/ragha/Downloads/13168478_3840_2160_25fps.mp4")

while True:
    ret , frame =  video.read()
    if not ret :
        break
    grey =  cv.cvtColor(frame , cv.COLOR_BGR2GRAY)
    plates = plate_cascades.detectMultiScale(grey , 1.1 , 10)
    for (x , y, w , h) in plates:
        cv.putText(frame , "Number Plate" , (x , y-7) ,cv.FONT_HERSHEY_SIMPLEX  , 4 , (0,255,0) , 3)
        cv.rectangle(frame , (x,y) ,(x+w , y+h) , (0,0,255) , 5)

    frame =  cv.resize(frame , (1000,500))
    cv.imshow("Result" , frame)
    
    plt.imshow(cv.cvtColor(frame , cv.COLOR_BGR2RGB))
    plt.axis("OFF")
    plt.show()

    if cv.waitKey(25) & 0xFF == ord('q'):
        break

video.release()
cv.destroyAllWindows()    
```
